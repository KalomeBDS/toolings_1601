```python
def list_min(lst):
    if not lst:
        raise ValueError("list_min() arg is an empty sequence")
    m = lst[0]
    for x in lst[1:]:
        if x < m:
            m = x
    return m

# Example
if __name__ == "__main__":
    values = [3, 1, 4, 0, -2]
    print(list_min(values))  # Output: -2
```
